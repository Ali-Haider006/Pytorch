***Manupliation***
Reshaping, Stacking, Squeezing and Unsqueezing
* Reshaping- changes shape of tensor
* View - shows a different shape of our input tensor
* Stack - Stack tensor vertically and horizontally (vstack and hstack)
* Squeeze- Removes one dimension
* Unsqeeze - add one dimension
* Permute - Return vieew of tensor with dimension permuted (swapped) in a certain way.

```python
x = torch.arange(1.,10.)
x, x.shape
```

New dimensions should always be compatable with the old dimenions.
* Rule of Thumb is the answer of multiplication of all new dimensions should be equal to number of elements in old tensor

```python
x.reshape(1,3,3)
```

View() is  same as reshape but, it shares same memory. It works like pointer. It will simply point towards x when we call z. Any changes to z will create change to x automatically.

```python
z = x.view(9,1)
z, z.shape
```

```python
z.view(9,1)
x
```

STACKING

```python
x = torch.arange(1,100.,10.)
x_stacked = torch.stack([x,x,x,x],dim=1)
x_stacked
```
